```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df=pd.read_csv('Order2.csv')
df
```

```python
ds=[]
for i in range(0,205):
    ds.append([str(df.values[i,j]) for j in range(0,5)])
```

```python
t1=apriori(ds,min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
t1_n=list(t1)
```

```python
print(len(t1_n))
```

```python
t1_n
```

```python
t2=list(t1_n)
```

```python
def inspect(t2):
    lhs         = [tuple(df[1][0])[0] for df in t2]
    rhs         = [tuple(df[1][0])[0] for df in t2]
    support    = [df[1] for df in t2]
    confidence = [df[1][0] for df in t2]
    lift       = [df[1][0] for df in t2]
    return list(zip(lhs, rhs, support, confidence, lift))
output_DataFrame = pd.DataFrame(inspect(df), columns = ['Left_Hand_Side', 'Right_Hand_Side', 'Support', 'Confidence', 'Lift'])
```
