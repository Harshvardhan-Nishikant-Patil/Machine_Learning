```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as mat
import seaborn as sn
from sklearn.model_selection import train_test_split 
#from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df=pd.read_csv('madfhantr.csv')
df
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
tmp=['Loan_ID','Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
```

```python
lbl=LabelEncoder()
for i in tmp:
    df[i]=lbl.fit_transform(df[i])
```

```python
x=df[['Gender','Married','Dependents','Education','Self_Employed','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']]
```

```python
y=df['Loan_Status']
```

```python
tr_in,tr_ot,ts_in,ts_ot=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ml=DecisionTreeClassifier().fit(tr_in,ts_in)
```

```python
ml.score(tr_ot,ts_ot)
```

```python
mat.figure(figsize=(144,81))
tree.plot_tree(ml,filled=True,fontsize=28)
mat.show()
```
